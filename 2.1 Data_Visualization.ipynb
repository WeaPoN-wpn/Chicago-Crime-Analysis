## Data Visualization

```python
import numpy as np
import seaborn as sns
import scipy
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\HKU\STAT4710\dataset_cleaned.csv')
df.head()
```

```python
# 过去十年每年犯罪数量的折线图，横坐标为年份，纵坐标为犯罪数量，每种颜色代表一种犯罪类型（UCR_PART_I和UCR_PART_II）
df['count'] = 1
df_count = df.groupby(['Year', 'UCR_PART']).size().reset_index(name='count')

color_dict = {'UCR_PART_I': 'C1', 'UCR_PART_II': 'C0'}

# 创建折线图
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='count', hue='UCR_PART', data=df_count, palette=color_dict)

# 设置 x 轴的刻度
plt.xticks(df_count['Year'].unique())

# 设置标题
plt.title('Number of Crimes by Year')

# 显示图像
plt.show()

```

```python
# 每个月的犯罪总数，横坐标为月份（每个月份一个颜色），纵坐标为犯罪数量
color_dict = {'UCR_PART_I': 'C1', 'UCR_PART_II': 'C0'}
plt.figure(figsize=(10, 5))
sns.countplot(x='Month', data=df, hue='UCR_PART', palette=color_dict)
plt.title('Number of Crimes by Month')
plt.show()
```

```python
# 每个Time的犯罪总数, 横坐标为Time，纵坐标为犯罪数量
color_dict = {'UCR_PART_I': 'C1', 'UCR_PART_II': 'C0'}
plt.figure(figsize=(7.5, 5))
sns.countplot(x='Time', data=df, hue='UCR_PART', palette=color_dict)
plt.title('Number of Crimes by Time')
plt.show()
```

```python
# Primary Type与UCR_PART的数量, 横坐标为Primary Type，纵坐标为犯罪数量, 每个UCR_PART一个颜色
color_dict = {'UCR_PART_I': 'C1', 'UCR_PART_II': 'C0'}
plt.figure(figsize=(10, 5))
sns.countplot(y='Primary Type', data=df, hue='UCR_PART', palette=color_dict)
plt.title('Number of Crimes by Primary Type')
plt.show()
```
